```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 3, 0.1)

def f(x):
    return 3 * x ** 2 - 4 * x

plt.plot(x, f(x), label='f(x)')
plt.plot(x, 2 * x - 3, label='Tangent line (x=1)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
